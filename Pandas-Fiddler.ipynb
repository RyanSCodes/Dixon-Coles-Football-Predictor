```python
import numpy as np
import pandas as pd

df = pd.read_csv('PremierLeague14.csv', dtype={'Date': str })
```

```python
df = df[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG'] ]
df.tail(50)
```

```python
date = "24/05/15"
day = date[:2]
month = date[3:5]
year = '20'+date[6:]
newdate = year + '-' + month + '-' + day
print newdate
```

```python
def reformat_date(date_string) :
    day = date_string[:2]
    month = date_string[3:5]
    year = "20"+ date_string[6:]
    newdate = year + "-" + month + '-' + day
    return newdate
```

```python
df['NewDate'] = df['Date'].apply(reformat_date)
df.head()
```

```python
df.NewDate = pd.to_datetime(df.NewDate)
df.tail(50)
```

```python
import datetime
#today = datetime.datetime.now().strftime("%Y-%m-%d")
today = datetime.datetime.now()
df['Now'] = today

def delta_time(times) :
    now, date = times
    A = np.datetime64(date)
    B = np.datetime64(now)
    C = (B-A)/(60*60*24*10**6)
    return C.astype(int)

df['DaysSince'] = df[['Now', 'Date']].apply(delta_time, axis=1)

#df = df[['DaysSince', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG']]

#results_list = df.as_matrix()

#results_list
```
